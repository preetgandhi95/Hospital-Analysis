```python
import pandas as pd
bs1 = pd.read_csv('bs1.csv')
bs1 = bs1.drop(bs1.columns[0], axis=1)

dates = {}
for patient in bs1['DESYNPUF_ID'][bs1['Y']==1]:
    hosp_dates = bs1[bs1['DESYNPUF_ID']==patient]['CLM_ADMSN_DT']
    min_date_2010 = min(hosp_dates[hosp_dates>=20100101])
    dates[patient] = min_date_2010
print(dates)
```

```python
bs2 = bs1.drop(['fst_admsn_dt','CLM_ADMSN_DT','DESYNPUF_ID','BENE_BIRTH_DT','BENE_DEATH_DT'],axis = 1)
```

```python

# Feature Importance with Extra Trees Classifier
from pandas import read_csv
from sklearn.ensemble import ExtraTreesClassifier
# load data

X = bs2.drop(['Y'],axis=1)
Y = bs2['Y']
# feature extraction
model = ExtraTreesClassifier()
model.fit(X, Y)
bs3 = {}
bs3['index']= bs2.columns
bs3['importance'] = model.feature_importances_
for x in bs3.keys():
    print(x)
    print(bs3[x])
```
